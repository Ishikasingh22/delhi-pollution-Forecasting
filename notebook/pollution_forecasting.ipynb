```python
# datetime column

data1['date'] = pd.to_datetime(
    dict(year=data1['Year'], 
         month=data1['Month'], 
         day=data1['Date'])
)
```

```python
data1[['Year','Month','Date','date']].head()
```

```python
data1.drop(['Year', 'Month', 'Date'], axis=1, inplace=True)

```

```python
data1 = data1.sort_values('date')
data1.set_index('date', inplace=True)
```

```python
#Selecting target variable

pm25 = data1[['PM2.5']]
```

```python
#Visualizing PM 2.5

pm25.plot(figsize=(12,5))
plt.title("Daily PM2.5 Levels")
plt.ylabel("PM2.5")
plt.show()
```

```python
pm25.to_csv("C:\\Users\\HP\\OneDrive\\Desktop\\pollution-level-forecasting\\data\\processed\\PM25_Processed.xlsx")
```

```python
# TEST-TRAIN split

train= pm25[:-365]
test = pm25[-365:]
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

decomposition = seasonal_decompose(train, model='additive', period=365)
decomposition.plot()
plt.show()
```

```python

plt.savefig("C:\\Users\\HP\\OneDrive\\Desktop\\pollution-level-forecasting\\images\\decomposition.png", dpi=300, bbox_inches='tight')
```

```python
#Checking for Stationarity

from statsmodels.tsa.stattools import adfuller

result = adfuller(train['PM2.5'])

print("ADF Statistic:", result[0])
print("p-value:", result[1])
```

```python
# testing the baseline model

forecast_baseline_naive = [train.iloc[-1]] * len(test)
```

```python
from sklearn.metrics import mean_absolute_error
import numpy as np

mae_naive = mean_absolute_error(test, forecast_baseline_naive)
rmse_naive = np.sqrt(((forecast_baseline_naive - test.values) ** 2).mean())

print("Naive MAE:", mae_naive)
print("Naive RMSE:", rmse_naive)
```

```python
forecast_baseline_seasonal = train[-365:].values
```

```python

forecast_baseline_seasonal_trimmed = forecast_baseline_seasonal[:len(test)]  

# Calculate metrics with matching dimensions
mae_seasonal = mean_absolute_error(test, forecast_baseline_seasonal_trimmed)
rmse_seasonal = np.sqrt(((forecast_baseline_seasonal_trimmed - test.values) ** 2).mean())

print("Seasonal Baseline MAE:", mae_seasonal)
print("Seasonal Baseline RMSE:", rmse_seasonal)

```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,5))
plt.plot(test.index, test, label="Actual")
plt.plot(test.index, forecast_baseline_seasonal_trimmed, label="Seasonal Baseline")
plt.legend()
plt.title("Daily Seasonal Baseline Forecast")
plt.show()
```

```python
plt.savefig("C:\\Users\\HP\\OneDrive\\Desktop\\pollution-level-forecasting\\images\\Forecast_Baseline.png", dpi=300, bbox_inches='tight')
```

```python
# Model is Stationary,building ARIMA

from statsmodels.tsa.arima.model import ARIMA

model = ARIMA(train, order=(1,0,1))
results = model.fit()

forecast_arima = results.forecast(steps=len(test))
```

```python
# Plotting the Forecast

import matplotlib.pyplot as plt

plt.figure(figsize=(12,5))
plt.plot(test, label='Actual')
plt.plot(forecast_arima, label='ARIMA Forecast')
plt.legend()
plt.show()
```

```python
plt.savefig("C:\\Users\\HP\\OneDrive\\Desktop\\pollution-level-forecasting\\images\\Forecast_ARIMA.png", dpi=300, bbox_inches='tight')
```

```python
from sklearn.metrics import mean_absolute_error

mae_arima = mean_absolute_error(test, forecast_arima)
print("ARIMA MAE:", mae_arima)
```

```python
# training and testing monthly aggregates

train = monthly_pm25[:-12]
test = monthly_pm25[-12:]
```

```python
# Visualizing

import matplotlib.pyplot as plt

plt.figure(figsize=(10,4))
plt.plot(train, label='Train')
plt.plot(test, label='Test')
plt.legend()
plt.title("Monthly PM2.5 - Train Test Split")
plt.show()
```

```python
# Checking for stationarity

from statsmodels.tsa.stattools import adfuller

result = adfuller(train)
print("ADF Statistic:", result[0])
print("p-value:", result[1])
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX

model = SARIMAX(
    train,
    order=(1,0,0),
    seasonal_order=(0,1,0,12),
    enforce_stationarity=False,
    enforce_invertibility=False
)

results = model.fit()
forecast_sarima = results.forecast(steps=12)
```

```python
plt.figure(figsize=(10,4))
plt.plot(train, label='Train')
plt.plot(test, label='Actual')
plt.plot(test.index, forecast_sarima, label='Forecast')
plt.legend()
plt.title("Monthly PM2.5 Forecast")
plt.show()
```
